```python
datos = {
    "nombre": ["Ana", "Juan", "Carla", "Pedro", "Lucia"],
    "nota": [8, 6, 9, 5, 7]
}

df = pd.DataFrame(datos)
print(df)
```

```python
suma=0
for i in df["nota"] :
    suma= suma + i
    print(i,"sumando lo anterior da",suma)
print ("La suma total de las notas es ", suma)
print ("El promedio de las notas es ", suma /len (df))

print ("EL PROMEDIO HECHO CON PANDA ES", df["nota"].mean())
```

```python
nota_maxima = df["nota"].max()
nota_minima = df["nota"].min()

print("La nota máxima con panda es", nota_maxima)
print("La Nota mínima con panda es", nota_minima)
```
